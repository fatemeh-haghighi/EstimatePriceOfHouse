# Guessing the value of w and b in linear equation


```python
def liniear_equation(w, x, b):
    return np.add(np.multiply(w, x), b)

def plot_liniear(x, y, s):
    plt.scatter(x , s , c= 'red')
    plt.scatter(x, y, c = 'blue')
    plt.ylabel('price')
    plt.xlabel('OverallQual')
    plt.show()
    
```

## Plotting real data and estimated linear equation in a window

```python
y = liniear_equation(40, no_NAN['OverallQual'], -80)
plot_liniear(no_NAN['OverallQual'],y ,no_NAN['SalePrice'])
```

# Calculating RMSE








```python
def Rmse(estimation, real, n):
    return np.power(np.divide((np.sum(np.power(estimation - real, 2))), n), 0.5)  
```

```python
print("RMSE output is: ", Rmse(y, no_NAN['SalePrice'], len(no_NAN.index)))
```

## standardizing dataframe by KNN algorithm


```python
def estimation(row):
    new_df = no_NAN.copy()
    prices = new_df['SalePrice']

    new_df.drop(['SalePrice', 'Id'], axis = 1, inplace = True)
    
#     standarding data
    min_val = new_df.min()
    max_val = new_df.max()
    def standardize(new_df, min_val, max_val):    
        return (new_df - min_val)/  (max_val - min_val)

    std_row = standardize(row, min_val, max_val)
    std = standardize(new_df, min_val, max_val)
#     print(std_row , std)
    
#     calculating distance
    dist = np.linalg.norm(std.values - std_row.values , axis = 1)
    
#      sorting distances
    k  = 10
    a = np.argpartition(dist, k)
    
#     calculating mean of k-th prices
    return(np.mean(prices[a[0: 10]]))
    
```

```python
sample = no_NAN[2:3]
sample = sample.drop(['Id', 'SalePrice'] , axis = 1)
print("Estimation of a house price is:", estimation(sample))
```

```python
rows = no_NAN.copy()
rows_prices = rows['SalePrice']

rows.drop(['SalePrice', 'Id'], axis = 1, inplace = True)
row_arr = rows.apply(estimation , axis=1)
print("Rmse of the whole dataframe is: ", Rmse(row_arr ,rows_prices , len(rows.index) ))
```

```python
plt.scatter(no_NAN['OverallQual'], no_NAN['SalePrice'],  color = 'blue', alpha = 0.4)
plt.scatter(no_NAN['OverallQual'], row_arr, color = 'red', alpha = 0.6)
plt.ylabel('price')
plt.xlabel('OverallQual')
plt.show()
```
